np.array():

Purpose: Creates a new NumPy array object from any object exposing the array interface.
Behavior: This function always creates a new array, copying the data from the input object. You can specify the desired data type and copy behavior using the dtype and copy parameters.
Use Case: Use np.array() when you need to ensure that you have a new array and possibly change the data type.

np.asarray():

Purpose: Converts the input to an array, but does not copy the data if the input is already an array or a compatible array-like object.
Behavior: This function will return the input itself if it is already a NumPy array. If not, it creates a new array only if necessary.
Use Case: Use np.asarray() when you want to avoid unnecessary copying and just need to ensure the input is converted to an array.

np.asanyarray():

Purpose: Similar to np.asarray(), but it allows for subclasses of ndarray, such as masked arrays.
Behavior: It will convert the input to an array, but if the input is already a subclass of ndarray, it returns that subclass without converting it to a standard ndarray.
Use Case: Use np.asanyarray() when working with objects that might be subclasses of ndarray and you want to preserve their type.

. Difference Between Deep Copy and Shallow Copy

Shallow Copy:

Definition: A shallow copy creates a new object, but it does not create copies of nested objects (i.e., it copies references to those objects).

Behavior: If you modify a nested object in the shallow copy, the change will reflect in the original object, as both share references to the same nested objects.



```python
import copy

original = [[1, 2, 3], [4, 5, 6]]
shallow_copied = copy.copy(original)  # Create a shallow copy

shallow_copied[0][0] = 'X'  # Modify nested object

print(original)
print(shallow_copied)
```

Deep Copy:

Definition: A deep copy creates a new object and recursively copies all objects found within the original object, creating independent copies of nested objects.

Behavior: Changes made to nested objects in the deep copy do not affect the original object, as they are entirely separate copies.

```python
import copy

original = [[1, 2, 3], [4, 5, 6]]
deep_copied = copy.deepcopy(original)  # Create a deep copy

deep_copied[0][0] = 'Y'  # Modify nested object

print(original)
print(deep_copied)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a date range
date_range = pd.date_range(start='2023-01-01', periods=100, freq='D')

# Step 2: Create random temperature and humidity data
np.random.seed(0)  # For reproducibility
temperature = np.random.uniform(low=15, high=35, size=len(date_range))  # Random temperature between 15 and 35
humidity = np.random.uniform(low=30, high=90, size=len(date_range))  # Random humidity between 30 and 90

# Step 3: Create a DataFrame
df = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})

# Step 4: Plotting
fig, ax1 = plt.subplots(figsize=(12, 6))

# a) Plot Temperature with the left y-axis
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.plot(df['Date'], df['Temperature'], color='tab:red', label='Temperature')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.plot(df['Date'], df['Humidity'], color='tab:blue', label='Humidity')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# b) Label the x-axis
ax1.set_xlabel('Date')

# c) Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Show the plot
plt.show()
```

17.Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate two random arrays
np.random.seed(0)  # For reproducibility
x = np.random.randn(100)  # 100 random values for x-axis
y = np.random.randn(100)  # 100 random values for y-axis

# Step 2: Create a DataFrame
df = pd.DataFrame({'X': x, 'Y': y})

# Step 3: Define quadrants
def get_quadrant(row):
    if row['X'] > 0 and row['Y'] > 0:
        return 'Quadrant I'
    elif row['X'] < 0 and row['Y'] > 0:
        return 'Quadrant II'
    elif row['X'] < 0 and row['Y'] < 0:
        return 'Quadrant III'
    elif row['X'] > 0 and row['Y'] < 0:
        return 'Quadrant IV'
    else:
        return 'Origin'

# Apply the function to determine quadrants
df['Quadrant'] = df.apply(get_quadrant, axis=1)

# Step 4: Create a scatter plot
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(data=df, x='X', y='Y', hue='Quadrant', palette='Set1', style='Quadrant', markers=['o', 's', 'D', '^'])

# Step 5: Label the axes and set the title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Add a legend
plt.legend(title='Quadrants')

# Show the plot
plt.grid()
plt.axhline(0, color='black',linewidth=0.5, ls='--')  # Add x-axis
plt.axvline(0, color='black',linewidth=0.5, ls='--')  # Add y-axis
plt.show()
```

18.With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine
Wave Function'.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

p = figure(title='Sine Wave Function', x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)
p.line(x, y, line_width=2, color='blue', legend_label='sin(x)')
p.grid.grid_line_alpha = 0.3

output_notebook()
show(p)
```

 19.Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 100, size=len(categories))

# Create a DataFrame
data = pd.DataFrame({'categories': categories, 'values': values})

# Create a ColumnDataSource
source = ColumnDataSource(data)

# Create a bar chart
p = figure(x_range=categories, title='Random Categorical Bar Chart',
           x_axis_label='Categories', y_axis_label='Values', width=800, height=400)

bars = p.vbar(x='categories', top='values', source=source,
              width=0.9, color='blue', legend_field='categories')

# Color bars based on values
p.vbar(x='categories', top='values', source=source,
       width=0.9, fill_color='blue', line_color='black')

# Add hover tooltips
hover = HoverTool(tooltips=[("Category", "@categories"), ("Value", "@values")])
p.add_tools(hover)

# Show the plot
output_notebook()
show(p)
```

20.Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as
'Simple Line Plot'
